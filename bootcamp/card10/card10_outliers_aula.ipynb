```python
%matplotlib inline
import numpy as np
incomes = np.random.normal(27000, 15000, 10000)
# com bilionario
incomes = np.append(incomes, [1000000000])

import matplotlib.pyplot as plt
print(f'\nRenda Média: {incomes.mean()}\n') # Valor da rendia média foi totalmente deturpado
plt.hist(incomes, 50)
plt.title(f"Distribuição de Renda Gerada\nRenda Média: R${round(incomes.mean(), 2)}")
plt.xlabel("Renda")
plt.ylabel("Frequência")
plt.show()
```

```python
def reject_outliers(data, m):
    u = np.median(data)
    s = np.std(data)
    filtered = [e for e in data if (u - m * s < e < u + m * s)]
    return filtered

filtered = reject_outliers(incomes, 2)  #(data, multiplo)
plt.hist(filtered, 50)
plt.title(f"Filtrando Outliers\nRenda Média: R${round(np.mean(filtered), 2)}")
plt.xlabel("Renda")
plt.ylabel("Frequência")
plt.show()
```

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

# Gerar dados normais
incomes = np.random.normal(27000, 15000, 10000)

# Adicionar vários outliers aleatórios
outliers = np.random.randint(1000000, 10000000, size=50)  # 50 outliers
incomes = np.append(incomes, outliers)

# Plotar o histograma original
plt.hist(incomes, 50)
plt.title("Histograma com Outliers")
plt.xlabel("Renda")
plt.ylabel("Frequência")
plt.show()

# Experimentar com diferentes múltiplos do desvio padrão
multiples = [1, 2, 3, 4]  # Diferentes valores de m
for m in multiples:
    filtered_incomes = reject_outliers(incomes, m)
    plt.hist(filtered_incomes, 50)
    plt.title(f"Histograma sem Outliers (m = {m})")
    plt.show()
    print(f"Número de dados após remoção de outliers (m = {m}): {len(filtered_incomes)}")
```
