## Perceptron 1 camada

```python
import numpy as np

# Função de ativação (degrau)
def step_function(x):
    return 1 if x >= 0 else 0

# Classe Perceptron
class Perceptron:
    def __init__(self, num_inputs):
        self.weights = np.zeros(num_inputs)
        self.bias = 0

    def predict(self, inputs):
        # Soma ponderada das entradas
        weighted_sum = np.dot(inputs, self.weights) + self.bias
        # Aplica a função de ativação
        return step_function(weighted_sum)

    def train(self, training_data, labels, learning_rate=0.1, epochs=100):
        for _ in range(epochs):
            for inputs, label in zip(training_data, labels):
                prediction = self.predict(inputs)
                # Atualiza pesos e bias
                self.weights += learning_rate * (label - prediction) * inputs
                self.bias += learning_rate * (label - prediction)

# Dados de treinamento para o operador "E"
training_data = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
labels = np.array([0, 0, 0, 1])

# Cria e treina o Perceptron
perceptron = Perceptron(num_inputs=2)
perceptron.train(training_data, labels)

# Testa o Perceptron
print("Testes do Perceptron para o operador 'E':")
for inputs in training_data:
    output = perceptron.predict(inputs)
    print(f"Entrada: {inputs} -> Saída: {output}")
```

```python
onehotenconder = ColumnTransformer(transformers=[('one_hot', OneHotEncoder(), [0, 2, 3, 8])], remainder='passthrough')
X = onehotenconder.fit_transform(X).toarray()
```

```python
X.shape
```

```python
camada_entrada = Input(shape=(303,))
camada_oculta1 = Dense(units=153, activation='relu')(camada_entrada)
camada_dropout = Dropout(rate=0.2)(camada_oculta1)
camada_oculta2 = Dense(units=153, activation='relu')(camada_dropout)
camada_saida1 = Dense(units=1, activation='linear')(camada_oculta2)

camada_saida2 = Dense(units=1, activation='linear')(camada_oculta2)
camada_saida3 = Dense(units=1, activation='linear')(camada_oculta2)
```

```python
regressor = Model(inputs=camada_entrada, outputs = [camada_saida1, camada_saida2, camada_saida3])
```

```python
regressor.compile(optimizer='adam', loss='mse', metrics=[['mean_absolute_error'], ['mean_absolute_error'], ['mean_absolute_error']]) # Provide a list of metrics for each output
```

```python
regressor.fit(X, [y_na, y_eu, y_jp], batch_size=100, epochs=200)
```

```python
prev_na, prev_eu, prev_jp = regressor.predict(X)
```
