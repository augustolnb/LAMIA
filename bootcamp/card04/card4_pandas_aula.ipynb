## Aula 33 - Concatenar, Juntar e Mesclar

df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                        'B': ['B0', 'B1', 'B2', 'B3'],
                        'C': ['C0', 'C1', 'C2', 'C3'],
                        'D': ['D0', 'D1', 'D2', 'D3']
                    })
                    

```python
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                        'B': ['B0', 'B1', 'B2', 'B3'],
                        'C': ['C0', 'C1', 'C2', 'C3'],
                        'D': ['D0', 'D1', 'D2', 'D3']},
                    index=[0, 1, 2, 3]
                    )
```

```python
df2 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                        'B': ['B0', 'B1', 'B2', 'B3'],
                        'C': ['C0', 'C1', 'C2', 'C3'],
                        'D': ['D0', 'D1', 'D2', 'D3']},
                    index=[0, 1, 2, 3]
                    )
```

```python
df3 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                        'B': ['B0', 'B1', 'B2', 'B3'],
                        'C': ['C0', 'C1', 'C2', 'C3'],
                        'D': ['D0', 'D1', 'D2', 'D3']},
                    index=[0, 1, 2, 3]
                    )
```

```python
df2
```

```python
# Concatena por padrão colunas de mesmo nome
pd.concat([df1, df2, df3], axis=1)

# Fazendo axis=1 a concatenação é feita por índice
```

```python
esquerda = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                     'A': ['A0', 'A1', 'A2', 'A3'],
                     'B': ['B0', 'B1', 'B2', 'B3']})

direita = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                     'C': ['C0', 'C1', 'C2', 'C3'],
                     'D': ['D0', 'D1', 'D2', 'D3']})
```

```python
esquerda
```

```python
direita
```

```python
# Mesclar
'''
    Mescla dataframes que tenham uma coluna em comum
'''

pd.merge(esquerda, direita, how='inner', on='key')


```

```python
esquerda.join(direita, lsuffix='_esquerda', rsuffix='_direita')
```
